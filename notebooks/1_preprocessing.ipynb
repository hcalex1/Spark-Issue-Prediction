# PREPROCESSING

```python
import pandas as pd
import seaborn as sns
```

```python
BUG_ISSUES_PATH = "../data/inputs/spark_bugs_3.0.0.csv"
GIT_LOGS_PATH = "../data/inputs/spark3.0_modified_files.csv"
SCALASTYLE_PATH = "../data/inputs/staticanalysis_output.csv"

SPARK_3_0_PREVIEW_RELEASE = pd.to_datetime("Oct 29, 2019") 
SPARK_3_0_RELEASE         = pd.to_datetime("Jun 05, 2020")
```

## Import Data

```python
bug_issues = pd.read_csv(BUG_ISSUES_PATH)
bug_issues.head()
```

```python
git_logs = pd.read_csv(GIT_LOGS_PATH)
git_logs.head()
```

```python
file_attributes = pd.read_csv(SCALASTYLE_PATH)
file_attributes.set_index("File", inplace=True)
file_attributes.head()
```

## Create Independent Variables

For the `file_attributes` DataFrame, lets remove all of the test files.

```python
file_attributes = file_attributes[file_attributes.index.str.contains("test")]
```

For the `bug_issues` DataFrame, lets set the *Issue key* column as the index, convert the *Created* column from strings to datetime objects and drop all the issues that were created before the release date of the Spark-3.0 preview.

```python
bug_issues = bug_issues.set_index("Issue key")
bug_issues["DTCreated"] = pd.to_datetime(bug_issues["Created"])
bug_issues = bug_issues[bug_issues["DTCreated"].map(lambda x : x >= SPARK_3_0_PREVIEW_RELEASE)]
```

Let take all the files and associate them with their first Jira bug issue and the date that this issue was created. We can drop all the files without an associated issues.

```python
issues = git_logs["Subject"].str.extractall("\[(SPARK-\d+)\]").sort_values(0)
issues["File"] = issues.index.map(lambda x: git_logs.loc[x[0], "File"])
issues["DTCreated"] = issues.loc[:,0].map(lambda x: bug_issues.loc[x, "DTCreated"] if (bug_issues.index == x).sum() else pd.NA)
issues.sort_values("DTCreated", inplace=True)
issues.drop_duplicates("File", keep='first', inplace=True)
issues.set_index("File", inplace=True)
```

Let create our independent variable by assining each of the files from the static analysis with a the time until the associated Jira bug issue is created, and take a look at the distribution of the `independent_variable` in weeks.

```python
time_to_first_bug = issues["DTCreated"].map(lambda t: t - SPARK_3_0_PREVIEW_RELEASE if type(t) == pd.Timestamp else pd.NaT)
independent_variable = time_to_first_bug.loc[time_to_first_bug.index.map(lambda i : i in file_attributes.index)]
(independent_variable.dt.days).plot(kind='hist', bins=50,title="Frequency distribution of the first bug report (in nb of days)")
```

## Correlation Analysis

```python
sns.heatmap(file_attributes.corr().abs(), vmin=0, vmax=1)
```

### Feature Engineering
Lets normalize the correlated features.

```python
file_attributes["LineCommentPerLineCode"]    = file_attributes["LineCommentCount"]     / file_attributes["LineCodeCount"]
file_attributes["LineCodeProportion"]        = file_attributes["LineCodeCount"]        / file_attributes["LinesCount"]
file_attributes["LineBlankProportion"]       = file_attributes["LineBlankCount"]       / file_attributes["LinesCount"]
file_attributes["ClassProportion"]           = file_attributes["ClassCount"]           / file_attributes["TypesCount"]
file_attributes["TraitProportion"]           = file_attributes["TraitCount"]           / file_attributes["TypesCount"]
file_attributes["ObjectProportion"]          = file_attributes["ObjectCount"]          / file_attributes["ClassCount"]
file_attributes["PrivateMethodProportion"]   = file_attributes["PrivateMethodCount"]   / file_attributes["MethodCount"]
file_attributes["ProtectedMethodProportion"] = file_attributes["ProtectedMethodCount"] / file_attributes["MethodCount"]
file_attributes["MethodOverrideProportion"]  = file_attributes["MethodOverrideCount"]  / file_attributes["MethodCount"]
file_attributes["FinalMethodProportion"]     = file_attributes["FinalMethodCount"]     / file_attributes["MethodCount"]
```

```python
features = file_attributes.loc[
    file_attributes.index.map(lambda i : i in independent_variable.index),
    ["LinesCount", "LinesAvgLength", "LineCommentPerLineCode", "LineCodeProportion", "LineBlankProportion", 
     "TypesCount", "ClassProportion", "TraitProportion", "ObjectProportion", "MaxChildren", 
     "MethodCount", "PrivateMethodProportion", "ProtectedMethodProportion", "MethodOverrideProportion", "FinalMethodProportion", 
     "VarLocalCount", "VarFieldCount", "MagicNumbersCount"]]
sns.heatmap(features.corr().abs(), vmin=0, vmax=1)
```

Lets check if there are any remaining features with correlations above 0.7 and remove them.

```python
sns.heatmap(features.corr().abs() >= 0.7)
```

```python
features.drop(["LineCommentPerLineCode", "ObjectProportion"], axis=1, inplace=True)
```

## Align the indices

```python
independent_variable = independent_variable[features.isna().sum(axis=1) == 0]
features.dropna(axis=0, inplace=True)

independent_variable = independent_variable.sort_index()
features = features.sort_index()
```

## Save Data

```python
features.to_csv("../data/output/features.csv")
independent_variable.to_csv("../data/output/time_to_bug.csv")
```
